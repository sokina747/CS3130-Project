##Summary of the Dataset

The dataset contains 20,000 rows and 54 columns describing individuals’ calorie intake, macro breakdowns, body measurements, and lifestyle habits. The main features used in my model were Calories, Fat_Percentage, BMI, protein_per_kg, and Daily meals frequency. These features had no missing values and showed meaningful variation needed for health classification.

##Key Findings from EDA

• Calories, carbs, and fats were right-skewed, with some individuals consuming very high amounts.
• Protein_per_kg tended to be higher in people with normal BMI and lower in individuals with overweight or obese BMI.
• BMI groups showed clear nutritional differences: healthier groups ate more balanced calories and had higher protein_per_kg.
• Calories from macros strongly correlated with total calories, and Fat_Percentage was a strong indicator of less healthy eating patterns.
• Overall, healthier individuals showed moderate calorie intake, higher protein_per_kg, and more balanced macro ratios.

##Model Insights

• All models (Decision Tree, Random Forest, and Tuned Random Forest) achieved very high accuracy on the test set, showing strong patterns in the data.
• Random Forest and Tuned Random Forest were the most stable models and reduced the risk of overfitting compared to a single Decision Tree.
• Feature importance analysis showed BMI, Fat_Percentage, Calories, and protein_per_kg as the strongest predictors of health status.
• Confusion matrices showed excellent class separation, indicating the models learned the healthy vs. unhealthy label effectively.
